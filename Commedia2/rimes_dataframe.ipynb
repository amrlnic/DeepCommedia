```
import string
import pandas as pd
import numpy as np
```

```
# load ascii text and covert to lowercase
filename = "Commedia_per_rime.txt"
f = open(filename, 'r', encoding='utf-8')
vector_text_temp = f.readlines()
vector_text = []
for el in vector_text_temp:
  temp = el.translate(str.maketrans('', '', string.punctuation))
  vector_text.append(temp)
```

```
print(vector_text)
```

```
words_X = []
words_y = []
new_cantica = True
i = 0
NEW_CANTO_FLAG = 'nuovocanto'

while i < len(vector_text)-3:

  if vector_text[i].split()[0] == NEW_CANTO_FLAG:
    new_cantica = True
    i += 1
    continue                                                                                                           

  if new_cantica:
    words_X.append(vector_text[i].split()[-1])
    words_y.append(vector_text[i+2].split()[-1])
    new_cantica = False
    i += 1

  elif vector_text[i+3].split()[0] == NEW_CANTO_FLAG:
    words_X.append(vector_text[i].split()[-1])
    words_y.append(vector_text[i+2].split()[-1])
    i += 3

  else:
    words_X.append(vector_text[i].split()[-1])
    words_X.append(vector_text[i+2].split()[-1])
    words_y.append(vector_text[i+2].split()[-1])
    words_y.append(vector_text[i+4].split()[-1])
    i += 3


```

```
for i in range(len(words_X)-100, len(words_X)):
  print("X: ",words_X[i], " y: ", words_y[i])
```

```
X = np.array(words_X)
y = np.array(words_y)

dict_df = {
            'X': X,
            'y': y
          }
words = np.vstack((X,y))
df = pd.DataFrame(dict_df)
```
